```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data structure: [restaurant_id, 2021,2022,2023,2024]
sales_data = np.array([[1,150000,180000,220000,250000], # paradise biryani
                       [2,120000,140000,160000,190000], # pizza hub
                       [3,200000,230000,260000,300000], # beijng bites
                       [4,180000,210000,240000,270000], # burger point 
                       [5,185000,160000,205000,230000]  # chai point
                       ])

print("=== tafa sales analysis === ")
print("\n sales data shape",sales_data.shape)
print("first three rest. sales are :",sales_data[:,1:])
print("first three rest. sales are :",sales_data[:3])
```

# total sales per year

```python
print(np.sum(sales_data,axis=0))
year_total= np.sum(sales_data[:,1:],axis=0)#1 for horizontal x-axis
# year_total= np.sum(sales_data[:,1:],axis=1)#2 for virtical y-axis
print(year_total)
print(np.sum(year_total,axis=0))
```

# minimum sales per restaurant

```python
min_sales=np.min(sales_data[:,1:],axis=1)#1 =row
                                         #0 =colum
print(min_sales)
```

# Maximum sales per year

```python
max_sales =np.max(sales_data[:,1:],axis=0)
print(max_sales)
```

# average sales per restautant

```python
avg_sales = np.mean(sales_data[:,1:],axis=1)
print(avg_sales)
```

```python
cumsum= np.cumsum(sales_data[:,1:],axis=1)
print(cumsum)
plt.Figure(figsize=(10,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Averg cumulative sales accross all restaurant")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```
